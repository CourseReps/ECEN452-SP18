## Background
Starting with a parallel plate capacitor, the capacitance is C0 when the region between the plates is a vacuum. If you replace the vacuum with an insulating material that effects the capacitance, that material is referred to a dielectric material. The factor that the dielectric material effects the capacitance is referred to as the dielectric constant, k. Dielectric materials always increase the capacitance, in other words the dielectric constant is always greater than 1. The relation is shown below in the figure I made using google drawing. ## Method
In this lab, the open coaxial probe method is utilized to measure the dielectric constant of a material, also referred to as the relative permittivity of a material. This method is depicted in the figure below. As shown in the figure below, the probe is pressed against the material or immersed in the liquid, the electric fields at the end of the probe fringes out into the material, this field reflects back. The reflected signal, i.e. the scattering parameter S11, can be related to the permittivity (Stutchly & Stutchly 1980). 
![pressed.png](pressed.png)


## Measurments
Prior to measuring a material, the system must be calibrated using standards for direct calibration at the open end of the probe. The standards used are an open, a short and a reference liquid, usually de-ionized water. The probe is very susceptible to air, to ensure proper measurement, check for air bubbles under the probe when its submersed into a liquid. Once calibration is complete, measurements can be taken of your material. The measurement files are exported using .csv format which can be used for analysis using python. 

## Data Analysis

```python
from matplotlib import pyplot as plt
import csv
```

```python
DI_Water = []
DI_water_20MHz_20GHz = []
Acetone = []
Ethylene_Glycol = []
Glass_Cleaner = []
Hydrocal = []
Isopropyl_Alcohol = []
Pine_sol = []
Silicone_Fluid = []
Simple_Green = []
WD_40 = []


freq = []
freq20 = []
```

```python

        
with open('Dielectric Values/DI_Water.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        freq.append(float(row['frequency']))
        DI_Water.append(float(row['e\'']))
        
with open('Dielectric Values/Acetone.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Acetone.append(float(row['e\'']))

with open('Dielectric Values/DI_water_20MHz_20GHz.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        freq20.append(float(row['frequency']))
        DI_water_20MHz_20GHz.append(float(row['e\'']))

with open('Dielectric Values/Ethylene_Glycol.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Ethylene_Glycol.append(float(row['e\'']))


with open('Dielectric Values/Glass_Cleaner.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Glass_Cleaner.append(float(row['e\'']))


with open('Dielectric Values/Hydrocal.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Hydrocal.append(float(row['e\'']))


with open('Dielectric Values/Isopropyl_Alcohol.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Isopropyl_Alcohol.append(float(row['e\'']))


with open('Dielectric Values/Pine_sol.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Pine_sol.append(float(row['e\'']))


with open('Dielectric Values/Silicone_Fluid.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Silicone_Fluid.append(float(row['e\'']))


with open('Dielectric Values/Simple_Green.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Simple_Green.append(float(row['e\'']))


with open('Dielectric Values/WD_40.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        WD_40.append(float(row['e\'']))
```

```python
plt.figure()  # initialize plot
ax1 = plt.subplot(111)
ax1.plot(freq, DI_Water, '-r', label="DI Water")
ax1.plot(freq, Glass_Cleaner, '--g', label="Glass Cleaner")
ax1.plot(freq, Ethylene_Glycol, '-k', label="Ethylene Glycol")
ax1.plot(freq, Isopropyl_Alcohol, '--b', label="Isopropyl Alcohol")
ax1.plot(freq, Hydrocal, '-y', label="Hydrocal")
ax1.plot(freq, Silicone_Fluid, '--c', label="Silicone Fluid")
ax1.plot(freq, WD_40, '-.y', label="WD-40")
ax1.plot(freq, Pine_sol, '-m', label="Pine Sol")
ax1.plot(freq, Simple_Green, '--c', label="Simple Green")
ax1.plot(freq, Acetone, '-g', label="Acetone")

ax1.set_xlim(min(freq), max(freq)) 
ax1.set_ylim(0, 90)  
ax1.legend(loc=4)  
plt.grid(b=True, which='both', color='0.65', linestyle='-')  
plt.title('Dielectric Constant - Liquids') 
plt.xlabel('Freq [GHz]') 
plt.tight_layout()
```
