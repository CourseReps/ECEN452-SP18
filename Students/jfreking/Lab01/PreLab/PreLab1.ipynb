### Add Parameters

Add a substrate to your microstrip line by double clicking on the line and selecting substrate. Because this is the first component of this design, make a new substrate now.



Choose the select option on the substrate definition to use a predefined material in the design. For now, select FR4 (listed as FR4_epoxy) for your material. Set H, your substrate thickness, to 62 mil, and HU to 1000 mm (do not set it to 1 meter -- this will cause frustrating errors when using the TRL to adjust your widths).

### Results

Now, view the results by adding a standard report. This is where you plot the S-parameters and Smith charts.

![preLabFig9.png](preLabFig9.png)

Create two separate reports. One should plot the magnitude of the S-parameters, and the other should plot their phase. Because this circuit consists of a single T-line, the S-parameters will be symmetric, so plotting S11 and S21 provides all the information needed. The magnitude plot is shown below.

![preLabFig10.png](preLabFig10.png)

The phase plot is shown here. Place a marker on the S21 plot and verify that there is about a 90 degree phase shift at the design frequency. Markers can be placed by right clicking on the plot. The location of the marker can be updated by double clicking the X or Y variables of the markers in the table.

![preLabFig11.png](preLabFig11.png)

### Export S-Parameters

Expand the Analysis folder on the left toolbar. Right click on LinearFrequency and select Export Solution...

![preLabFig12.png](preLabFig12.png)

Select Export Matrix Data... and save it as a .sNp file. Save the file in a known location to access it later. Keep all default parameters.
