Finally, a comparison of these tests was done with scikit-rf:

```python
import skrf as rf
import numpy as np
from matplotlib import pyplot as plt

gnd = rf.Network('Amplifier_Grounded.s2p')
radStub60 = rf.Network('Amplifier_RadialStub_60deg.s2p')
radStub90 = rf.Network('Amplifier_RadialStub_90deg.s2p')
```

```python
gnd.s21.plot_s_db(label="RF Shorted")
radStub60.s21.plot_s_db(label="RF 60deg Radial Stub")
radStub90.s21.plot_s_db(label="RF 90deg Radial Stub")
plt.grid()
plt.title("Comparison of Gains")
```

```python
gnd.s11.plot_s_db(label="RF Shorted")
radStub60.s11.plot_s_db(label="RF 60deg Radial Stub")
radStub90.s11.plot_s_db(label="RF 90deg Radial Stub")
plt.grid()
plt.title("Comparison of Input Matches")
```

### Conclusions

The ground termination resulted in better performance at lower frequencies, but the gain and input matching degraded at higher frequencies. The radial stubs performed better at high frequencies and the angle of aperture had little impact on the performance.
